```python

import yfinance as yf
from datetime import date
import dateutil.relativedelta

class TickerData():

  def getData(tickerSymbol, periodSize, periods):

    #get data on this ticker
    pulledData = yf.Ticker(tickerSymbol)

    #return the historical prices for this ticker
    return pulledData.history(period=periodSize, start=date.today() - dateutil.relativedelta.relativedelta(days=periods), end=date.today().strftime("%Y-%m-%d"))

#Call the Class in bult function
TickerData.getData('MSFT', '1d', 3650)
```
